```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import warnings
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
from scratch_decision_tree import ScratchDecesionTreeClassifierDepth2
from node import Node2
from information_gain import information_gain2
warnings.simplefilter("ignore")
```

```python
#irisデータのデータ50個、特徴量３つ
data = load_iris()
X3 = data["data"][25:75,1:4]
y3 = data["target"][25:75]
train_x,test_x,train_y,test_y = train_test_split(X3,y3,train_size=0.8,random_state=0)
```

```python
dtc = ScratchDecesionTreeClassifierDepth2()
dtc.fit(train_x,train_y)
y_pred1 = dtc.predict(test_x)
print("正解率：",accuracy_score(test_y,y_pred1))
```
